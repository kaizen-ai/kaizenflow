```python
import requests
import pandas as pd
```

```python
url = "https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest"
headers = {
    "Accepts": "application/json",
    "X-CMC_PRO_API_KEY": "49f9d121-9924-4b11-84b8-9b49b3497676",
}
# The API only allows us to get at max 5000 coins back as a response.
parameters = {
    "limit": "5000",
    "convert": "USD"
}

response = requests.get(url, headers=headers, params=parameters)

data = response.json()["data"]
df = pd.json_normalize(data, sep="_")
df
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)
df
```

```python
kept_columns = ['name', 'slug', 'quote_USD_price', 'quote_USD_volume_24h', 'quote_USD_volume_change_24h', \
                'quote_USD_percent_change_1h', 'quote_USD_percent_change_24h', 'quote_USD_percent_change_7d', \
                'quote_USD_percent_change_30d',	'quote_USD_percent_change_60d',	'quote_USD_percent_change_90d', \
                'quote_USD_market_cap']
data = df[kept_columns]
data.rename(columns = {
    'quote_USD_price': 'price',
    'quote_USD_volume_24h': 'volume_24h',
    'quote_USD_volume_change_24h': 'volume_change_24h',
    'quote_USD_percent_change_1h': 'percent_change_1h',
    'quote_USD_percent_change_24h': 'percent_change_24h',
    'quote_USD_percent_change_7d': 'percent_change_7d',
    'quote_USD_percent_change_30d': 'percent_change_30d',
    'quote_USD_percent_change_60d': 'percent_change_60d',
    'quote_USD_percent_change_90d': 'percent_change_90d',
    'quote_USD_market_cap': 'market_cap'
})
```
