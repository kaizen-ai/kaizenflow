# Bare CCXT extractor test

```python
import ccxt

extractor = ccxt.okx()
```

```python
start_timestamp = pd.Timestamp("2020-01-01T00:00:00+00:00")
start_timestamp = hdateti.convert_timestamp_to_unix_epoch(start_timestamp)
currency_pair = "ETH/USDT"
timeframe = "1m"
limit = 500
```
