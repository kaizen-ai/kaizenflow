```python
import pyarrow as pa
import pyarrow.dataset as ds
import pyarrow.parquet as pq
```

```python
file_name = "/data/projects/SorrIssue2_Cross_exchange_arbitrage_CEX_CEX/OKX_futures"
partitioning = ds.partitioning(schema = None, flavor="hive")
dataset = pq.ParquetDataset(
    # Replace URI with path.
    file_name,
    #filesystem=filesystem,
    #filters=filters,
    partitioning=partitioning,
    use_legacy_dataset=False,
)
```

```python
columns = None
table = dataset.read_pandas(columns=columns)
df_Binance_Spot = table.to_pandas()
df_Binance_Spot
```

```python
df_Binance_Spot.to_csv('df_binanace_spot.csv')
```
