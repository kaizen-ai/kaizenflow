```python
import pytz
import pandas as pd

import sorrentum_sandbox.examples.ml_projects.Issue22_Team3_Implement_sandbox_for_Coinmarketcap as coinmarketcap

coinmarketcap_load = coinmarketcap.load_data_from_db
```

```python
bitcoin_features_data=coinmarketcap_load.get_quote_data('bitcoin_features')
bitcoin_features_data = bitcoin_features_data.loc[bitcoin_features_data['last_updated'] > '2023-05-01']
bitcoin_features_data
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

bitcoin_data = bitcoin_features_data

# Convert type of column last_updated to datetime
bitcoin_data['last_updated'] = pd.to_datetime(bitcoin_data['last_updated'])

# Set last_updated as index
bitcoin_data.set_index('last_updated', inplace=True)

# draw price
bitcoin_data['price'].plot()

# Set title, xlabel, ylable for graph
plt.title('Price')
plt.xlabel('Last Updated Time')
plt.ylabel('Price')

# Show graph
plt.show()
```

```python
# draw nvt_ratio graph
bitcoin_data['nvt_ratio'].plot()

# Set title, xlabel, ylable for graph
plt.title('NVT Ratio')
plt.xlabel('Last Updated Time')
plt.ylabel('NVT Ratio')

# Show graph
plt.show()
```

```python
# draw MC/FDMC graph
bitcoin_data['mc_fdmc_ratio'].plot()

# Set title, xlabel, ylable for graph
plt.title('MC/FDMC')
plt.xlabel('Last Updated Time')
plt.ylabel('MC/FDMC')

# Show graph
plt.show()
```
