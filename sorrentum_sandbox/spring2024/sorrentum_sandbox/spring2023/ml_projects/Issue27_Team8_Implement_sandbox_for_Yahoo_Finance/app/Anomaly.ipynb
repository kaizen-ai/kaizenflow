```python
from sklearn.neighbors import  LocalOutlierFactor
```

```python
from api import api_from_db
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
real_time_data=api_from_db("2023-04-07","2023-04-16",'1m',['AVAX-USD'])
historic_time_data=api_from_db("2023-03-01","2023-04-16",'1d',['AVAX-USD'])

historic_time_data['open']=historic_time_data['open'].astype('float')
historic_time_data['high']=historic_time_data['high'].astype('float')
historic_time_data['low']=historic_time_data['low'].astype('float')
historic_time_data['close']=historic_time_data['close'].astype('float')
historic_time_data['adj_close']=historic_time_data['adj_close'].astype('float')
historic_time_data['volume']=historic_time_data['volume'].astype('float')

```

```python
real_time_data['open']=real_time_data['open'].astype('float')
real_time_data['high']=real_time_data['high'].astype('float')
real_time_data['low']=real_time_data['low'].astype('float')
real_time_data['close']=real_time_data['close'].astype('float')
real_time_data['adj_close']=real_time_data['adj_close'].astype('float')
real_time_data['volume']=real_time_data['volume'].astype('float')

```

```python
historic_time_data['close']=historic_time_data['close'].astype('float')
model = LocalOutlierFactor(n_neighbors=20, contamination=0.05)  # Adjust the parameters as needed
anomaly_scores = model.fit_predict(historic_time_data[['close']])
historic_time_data['outlier'] = anomaly_scores
```

```python
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)

x = list(historic_time_data['timestamp'])
y = list(historic_time_data['close'])

# function below sets the color based on amount
def SetColor(x):
    if(x == 1):
        return "green"
    if (x==-1):
        return "red"
    


# Create a trace
trace =  go.Scatter(
    x = x,
    y = y,
    mode='markers',
    marker_color =historic_time_data['outlier']
)

iplot([trace], filename='basic-line')
```
